```python
import pandas as pd

# ============================================================
# VALIDAÇÃO COMPLETA — 3 SETORES
# ============================================================

# Carregar todos os KPIs e Z-Scores
kpis_fin = pd.read_csv("../KPIs/kpis_financeiro.csv", sep=";", encoding="utf-8", keep_default_na=False)
kpis_en  = pd.read_csv("../KPIs/kpis_energia.csv",    sep=";", encoding="utf-8", keep_default_na=False)
kpis_pet = pd.read_csv("../KPIs/kpis_petroleo.csv",   sep=";", encoding="utf-8", keep_default_na=False)

zscores_fin = pd.read_csv("../KPIs/zscores_financeiro.csv", sep=";", encoding="utf-8", keep_default_na=False)
zscores_en  = pd.read_csv("../KPIs/zscores_energia.csv",    sep=";", encoding="utf-8", keep_default_na=False)
zscores_pet = pd.read_csv("../KPIs/zscores_petroleo.csv",   sep=";", encoding="utf-8", keep_default_na=False)

# ============================================================
# BLOCO 1 — Cobertura por setor e ano
# ============================================================
print("=" * 60)
print("BLOCO 1 — COBERTURA POR SETOR E ANO")
print("=" * 60)

for nome, df in [('FINANCEIRO', kpis_fin), ('ENERGIA', kpis_en), ('PETRÓLEO', kpis_pet)]:
    print(f"\n--- {nome} ---")
    print(df.groupby(['ano', 'tipo'])['empresa'].count().unstack(fill_value=0).to_string())

# ============================================================
# BLOCO 2 — Nulos por setor
# ============================================================
print("\n" + "=" * 60)
print("BLOCO 2 — NULOS POR SETOR")
print("=" * 60)

for nome, df in [('FINANCEIRO', kpis_fin), ('ENERGIA', kpis_en), ('PETRÓLEO', kpis_pet)]:
    print(f"\n--- {nome} ---")
    for col in df.columns:
        vazios = df[col].isin(['', 'None']).sum() + df[col].isna().sum()
        if vazios > 0:
            print(f"  {col}: {vazios} vazios")

# ============================================================
# BLOCO 3 — Estatísticas descritivas por setor
# ============================================================
print("\n" + "=" * 60)
print("BLOCO 3 — ESTATÍSTICAS DESCRITIVAS")
print("=" * 60)

kpis_fin_num = ['roe', 'margem_liquida', 'nim', 'indice_eficiencia', 'alavancagem', 'pdd_ratio']
kpis_en_num  = ['roe', 'margem_liquida', 'margem_bruta', 'margem_ebitda', 'alavancagem', 'cobertura_juros']
kpis_pet_num = ['roe', 'margem_liquida', 'margem_bruta', 'margem_ebitda', 'alavancagem', 'cobertura_juros']

for nome, df, cols in [
    ('FINANCEIRO', kpis_fin, kpis_fin_num),
    ('ENERGIA',    kpis_en,  kpis_en_num),
    ('PETRÓLEO',   kpis_pet, kpis_pet_num),
]:
    print(f"\n--- {nome} ---")
    print(df[cols].describe().round(4).to_string())

# ============================================================
# BLOCO 4 — Transições e eventos exógenos
# ============================================================
print("\n" + "=" * 60)
print("BLOCO 4 — TRANSIÇÕES E EVENTOS EXÓGENOS")
print("=" * 60)

# Eletrobras
print("\n--- ELETROBRAS ---")
print(kpis_en[kpis_en['empresa'] == 'CENTRAIS ELET BRAS S.A. - ELETROBRAS'][
    ['ano', 'tipo', 'evento_exogeno', 'descricao_evento']
].to_string(index=False))

# Copel
print("\n--- COPEL ---")
print(kpis_en[kpis_en['empresa'] == 'CIA PARANAENSE DE ENERGIA - COPEL'][
    ['ano', 'tipo', 'evento_exogeno', 'descricao_evento']
].to_string(index=False))

# Petrobras
print("\n--- PETROBRAS ---")
print(kpis_pet[kpis_pet['empresa'] == 'PETROLEO BRASILEIRO S.A. PETROBRAS'][
    ['ano', 'tipo', 'evento_exogeno', 'descricao_evento']
].to_string(index=False))

# ============================================================
# BLOCO 5 — Z-Scores das estatais ao longo dos anos
# ============================================================
print("\n" + "=" * 60)
print("BLOCO 5 — Z-SCORES DAS ESTATAIS POR ANO")
print("=" * 60)

# BB — Financeiro
print("\n--- BANCO DO BRASIL (ROE e Margem Líquida) ---")
bb = zscores_fin[zscores_fin['empresa'] == 'BCO BRASIL S.A.']
print(bb[bb['kpi'].isin(['roe', 'margem_liquida'])][
    ['ano', 'kpi', 'valor', 'media_privados', 'zscore', 'sinal']
].to_string(index=False))

# Cemig — Energia
print("\n--- CEMIG (ROE e Margem EBITDA) ---")
cemig = zscores_en[zscores_en['empresa'] == 'CIA ENERGETICA DE MINAS GERAIS - CEMIG']
print(cemig[cemig['kpi'].isin(['roe', 'margem_ebitda'])][
    ['ano', 'kpi', 'valor', 'media_privados', 'zscore', 'sinal']
].to_string(index=False))

# Petrobras — Petróleo
print("\n--- PETROBRAS (ROE e Margem EBITDA) ---")
pet = zscores_pet[zscores_pet['empresa'] == 'PETROLEO BRASILEIRO S.A. PETROBRAS']
print(pet[pet['kpi'].isin(['roe', 'margem_ebitda'])][
    ['ano', 'kpi', 'valor', 'media_privados', 'zscore', 'sinal']
].to_string(index=False))
```

```python
import pandas as pd

# ============================================================
# INTEGRAR MACRO COM KPIs
# Verificar se o ROE das estatais supera a Selic Real
# ============================================================

# Carregar dados
macro    = pd.read_csv("../Dados_macroeconomicos/macro_anual.csv", sep=";", encoding="utf-8")
kpis_fin = pd.read_csv("../KPIs/kpis_financeiro.csv", sep=";", encoding="utf-8", keep_default_na=False)
kpis_en  = pd.read_csv("../KPIs/kpis_energia.csv",    sep=";", encoding="utf-8", keep_default_na=False)
kpis_pet = pd.read_csv("../KPIs/kpis_petroleo.csv",   sep=";", encoding="utf-8", keep_default_na=False)

# Unir todos os KPIs em um único dataframe
kpis_todos = pd.concat([
    kpis_fin[['empresa', 'setor', 'tipo', 'ano', 'roe']],
    kpis_en[['empresa',  'setor', 'tipo', 'ano', 'roe']],
    kpis_pet[['empresa', 'setor', 'tipo', 'ano', 'roe']],
], ignore_index=True)

# Cruzar com macro pelo ano
kpis_macro = kpis_todos.merge(
    macro[['ano', 'ipca_anual', 'selic_anual', 'selic_real']],
    on='ano'
)

# Converter Selic Real de % para decimal para comparar com ROE
kpis_macro['selic_real_dec'] = kpis_macro['selic_real'] / 100

# Calcular se ROE supera a Selic Real
# True = empresa cria valor real acima do custo de oportunidade
kpis_macro['bate_selic_real'] = kpis_macro['roe'] > kpis_macro['selic_real_dec']

# Calcular diferença entre ROE e Selic Real (spread)
kpis_macro['spread_selic'] = (kpis_macro['roe'] - kpis_macro['selic_real_dec']).round(4)

# Ver só as estatais
estatais = kpis_macro[kpis_macro['tipo'] == 'Estatal']


print("=== ESTATAIS — ROE vs SELIC REAL ===\n")
print(estatais[['empresa', 'ano', 'setor', 'roe', 'selic_real_dec', 'spread_selic', 'bate_selic_real']]
      .sort_values(['empresa', 'ano'])
      .to_string(index=False))
```

```python
# ============================================================
# SALVAR RESULTADO INTEGRADO — KPIs + MACRO
# ============================================================

import os


os.makedirs("../Dados_macroeconomicos", exist_ok=True)

# Salvar todas as empresas (estatais e privadas) com macro
kpis_macro.to_csv(
    "../Dados_macroeconomicos/kpis_macro_integrado.csv",
    index=False,
    sep=";",
    encoding="utf-8",
    na_rep=''
)

# Salvar só estatais para análise focada
estatais.to_csv(
    "../Dados_macroeconomicos/estatais_vs_selic.csv",
    index=False,
    sep=";",
    encoding="utf-8",
    na_rep=''
)

print("Arquivos salvos!")
print(f"Total integrado: {len(kpis_macro)} linhas")
print(f"Só estatais: {len(estatais)} linhas")
```
