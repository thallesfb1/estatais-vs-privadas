```python
import pandas as pd

kpis_en = pd.read_csv("../KPIs/kpis_energia.csv", sep=";", encoding="utf-8", keep_default_na=False)
kpis_fin = pd.read_csv("../KPIs/kpis_financeiro.csv", sep=";", encoding="utf-8", keep_default_na=False)


print("=== FINANCEIRO — empresas por ano ===")
print(kpis_fin.groupby(['ano', 'tipo'])['empresa'].count().unstack(fill_value=0))

print("\n=== ENERGIA — empresas por ano ===")
print(kpis_en.groupby(['ano', 'tipo'])['empresa'].count().unstack(fill_value=0))
```

```python
print("=== NULOS FINANCEIRO ===")
print(kpis_fin.isnull().sum())

print("\n=== NULOS ENERGIA ===")
print(kpis_en.isnull().sum())
```

```python
kpis_fin_num = ['roe', 'margem_liquida', 'nim', 'indice_eficiencia', 'alavancagem', 'pdd_ratio']
kpis_en_num  = ['roe', 'margem_liquida', 'margem_bruta', 'margem_ebitda', 'alavancagem', 'cobertura_juros']

print("=== ESTATÍSTICAS FINANCEIRO ===")
print(kpis_fin[kpis_fin_num].describe().round(2))

print("\n=== ESTATÍSTICAS ENERGIA ===")
print(kpis_en[kpis_en_num].describe().round(2))
```

```python
print("=== ELETROBRAS — tipo por ano ===")
print(kpis_en[kpis_en['empresa'] == 'CENTRAIS ELET BRAS S.A. - ELETROBRAS'][['ano', 'tipo', 'evento_exogeno', 'descricao_evento']].to_string(index=False))

print("\n=== COPEL — tipo por ano ===")
print(kpis_en[kpis_en['empresa'] == 'CIA PARANAENSE DE ENERGIA - COPEL'][['ano', 'tipo', 'evento_exogeno', 'descricao_evento']].to_string(index=False))
```

```python
zscores_fin = pd.read_csv("../KPIs/zscores_financeiro.csv", sep=";", encoding="utf-8")
zscores_en  = pd.read_csv("../KPIs/zscores_energia.csv",    sep=";", encoding="utf-8")

print("=== Z-SCORES FINANCEIRO — BB ao longo dos anos ===")
bb = zscores_fin[zscores_fin['empresa'] == 'BCO BRASIL S.A.'].pivot(index='ano', columns='kpi', values='zscore')
print(bb.round(2).to_string())

print("\n=== Z-SCORES ENERGIA — CEMIG ao longo dos anos ===")
cemig = zscores_en[zscores_en['empresa'] == 'CIA ENERGETICA DE MINAS GERAIS - CEMIG'].pivot(index='ano', columns='kpi', values='zscore')
print(cemig.round(2).to_string())
```
