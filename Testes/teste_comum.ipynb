```python
from bcb import sgs
import pandas as pd
```

```python
dados = sgs.get(
    {"selic_mes": 4390, "ipca_mes": 433},
    start="2014-01-01",
    end="2023-12-31"
)

# Agrupar por ano e acumular
resultado = dados.resample("YE").apply(
    lambda x: (((1 + x / 100).prod()) - 1) * 100
)
print(resultado)
resultado.round(2)
resultado.to_csv("dados_macroeconomicos.csv", encoding="utf-8", sep=";", float_format="%.2f")
```
